```python
import pandas as pd
df = pd.read_csv("AirQualityUCI.csv", sep=";", decimal=",")
df.head()
```

```python
df.info()
```

```python
# df=df.drop(columns=['Unnamed: 15', 'Unnamed: 16', 'Date', 'Time'],axis=1)
df
```

```python
df.dropna(inplace=True)
```

```python

df['NMHC(GT)']= df['NMHC(GT)'].replace(-200, np.nan)
# df.isna().sum()
df
```

```python
df.isna().sum()
```

```python
df = df.drop(columns = ['NMHC(GT)'], axis = 1)
```

```python
y = df["AH"]
x = df.drop('AH', axis = 1)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import Ridge

X_train,X_test,Y_train,Y_test = train_test_split(x, y, test_size=0.2, random_state=56)
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)

ri = Ridge(alpha=12)
ri.fit(X_train, Y_train)

y_pred = ri.predict(X_test)
print(ri.score(X_test, Y_test))
```

```python
from sklearn.linear_model import Lasso

la  = Lasso()
la.fit(X_train, Y_train)

y_pred2 = la.predict(X_test)
print(la.score(X_test, Y_test))
```
