```python
import numpy as np
def med(df):
  l0 = []
  l1 = []
  for i in range(len(df['BMI'])):
    if df['OUTCOME'][i] == 0:
      l0.append(df['BMI'][i])
    elif df['OUTCOME'][i] == 1:
      l1.append(df['BMI'][i])
  # print(l0)
  median_l0 = np.median(l0)
  median_l1 = np.median(l1)
  print('Median for 0 outcome:',median_l0)
  print('Median for 1 outcome:',median_l1)

med(data)
data.head()
```

```python
data['AGE_CATEGORY'] = data['AGE'].apply(lambda x: 'Young' if x == 21 else 'Middle_Young' if x >21 and x<50 else 'Old')
data['AGE_CATEGORY']
```

```python
from typing import List
import numpy as np


def replace_zero(df, columns:List[str]):
    for column in columns:
        data[column] = data[column].replace(0, np.NaN)
    return df
data = replace_zero(data,['BMI','GLUCOSE','BLOODPRESSURE','SKINTHICKNESS','INSULIN'])
data['BLOODPRESSURE'].fillna(data['BLOODPRESSURE'].mean(),inplace = True)
data

```
