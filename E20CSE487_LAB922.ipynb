```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/adult.data", header=None)
df.head()
```

```python
import numpy as np
```

```python

df.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education_num', 'marital_status',
              'occupation', 'relationship', 'race', 'sex', 'capital_gain', 'capital_loss', 
              'hours_per_week', 'native_country','churn']
df.replace(" ?", np.NaN, inplace=True)
df.isnull().sum()
```

```python
mode_workclass = df['workclass'].mode()
mode_occupation = df['occupation'].mode()
mode_native_country = df['native_country'].mode()
print("Mode for Workclass -->",mode_workclass)
print("Mode for Occupation -->", mode_occupation)
print("Mode for Native-Country -->", mode_native_country)
```

```python
# Handling all missing values

df['workclass'].replace(np.nan, 'Private', inplace=True)
df['occupation'].replace(np.nan, 'Prof-specialty', inplace=True)
df['native_country'].replace(np.nan, 'United-States', inplace=True)


df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cato = ['sex', 'workclass', 'relationship','race','marital_status','education','churn']
for i in cato:
  df[i] = le.fit_transform(df[i])
df=df.drop(columns=['occupation','native_country'])
df.head()
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
```

```python
sl = StandardScaler()
Y = df['churn']
X = df.drop('churn', axis=1)

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=52)

X_train = sl.fit_transform(X_train)
X_test = sl.transform(X_test)
```

```python
gbc = GradientBoostingClassifier(n_estimators=100, learning_rate=0.3)
# gbc = GradientBoostingClassifier(estimators=[("svc", svc), ('lr', lr), ('dtc',dtc)])

gbc.fit(X_train, Y_train)
Y_pred = gbc.predict(X_test)

from sklearn.metrics import accuracy_score
accuracy_score(Y_test, Y_pred)*100
```

```python
from sklearn.metrics import roc_curve, auc
import seaborn as sns
roc = roc_curve(Y_test, gbc.decision_function(X_test))
res = auc(roc[0], roc[1])

sns.lineplot(x=roc[0], y=roc[1])
```

```python
from sklearn.metrics import accuracy_score
acc =[]
lrs = [0.05,0.1,0.2,0.3,0.35,0.4,0.5,0.6,0.7,0.8,0.9]
for i in lrs:
  gbc = GradientBoostingClassifier(n_estimators=100, learning_rate=i)
  gbc.fit(X_train, Y_train)
  Y_pred = gbc.predict(X_test)
  acc.append(accuracy_score(Y_test, Y_pred))
  print("LR:",i,"Accuracy:",accuracy_score(Y_test, Y_pred)*100)
  
```

```python
sns.lineplot(x=lrs, y=acc)
```
