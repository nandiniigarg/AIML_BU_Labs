```python
species=df['Species'].value_counts()
print(species.index)
print(species.values)
plt.bar(x=species.index,height=species.values,color=['blue','green','red'])
```

```python
plt.plot(df.Id,df["SepalLengthCm"],linestyle="dashed")
```

```python
x=df['SepalLengthCm']
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,x,z)
plt.xlabel('x values from 0 to 4pi')
plt.ylabel('sin(x) and cos(x)')
plt.title('Plot of sin and cos from 0 to 4pi')
plt.legend(['sin(x)','cos(x)'])
plt.show()
```
